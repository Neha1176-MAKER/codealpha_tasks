```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")

# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year
df['year_added'] = df['date_added'].dt.year

# Count content by year
content_by_year = df['year_added'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10,5))
plt.plot(content_by_year.index, content_by_year.values, marker='o')
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.title("Netflix Content Added Per Year")
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df)
plt.title('Movies vs TV Shows on Netflix')
plt.show()
```

```python
type_year = df.groupby(['year_added','type']).size().reset_index(name='count')

plt.figure(figsize=(10,5))
sns.lineplot(data=type_year, x='year_added', y='count', hue='type')
plt.title('Movies vs TV Shows Trend Over Time')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(y=top_countries.index, x=top_countries.values)
plt.title('Top 10 Content Producing Countries on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
genres = df['listed_in'].str.split(',').explode().str.strip()
top_genres = genres.value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(y=top_genres.index, x=top_genres.values)
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```
